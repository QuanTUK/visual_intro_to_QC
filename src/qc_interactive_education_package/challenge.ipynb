```python
import os
import json
import qiskit.qasm2
from qc_interactive_education_package import ChallengeViewer

try:
    num_qubits = int(os.environ.get("CHALLENGE_QUBITS", 1))
    initial_state = json.loads(os.environ.get("CHALLENGE_INITIAL", "[1, 0]"))
    target_state = json.loads(os.environ.get("CHALLENGE_TARGET", "[1, -1]"))

    show_circuit_flag = os.environ.get("CHALLENGE_SHOW_CIRCUIT", "1")
    show_circuit = True if show_circuit_flag == "1" else False

    preloaded_qasm = os.environ.get("CHALLENGE_PRELOADED_QASM", None)
    preloaded_circuit = None
    if preloaded_qasm:
        preloaded_circuit = qiskit.qasm2.loads(preloaded_qasm)

except Exception as e:
    print(f"Error parsing environment variables. Details: {e}")
    num_qubits = 1
    initial_state = [1, 0]
    target_state = [1, -1]
    show_circuit = True
    preloaded_circuit = None

viewer = ChallengeViewer(
    num_qubits=num_qubits,
    initial_state=initial_state,
    target_state=target_state,
    preloaded_circuit=preloaded_circuit,
    show_circuit=show_circuit
)
viewer.display()
```
