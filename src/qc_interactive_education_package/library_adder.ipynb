```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFTGate, grover_operator
from qiskit.quantum_info import Statevector
from qc_interactive_education_package import LibraryManager

# 1. Instantiate the manager
db = LibraryManager()

# ==========================================
# Add an Algorithm: 3-Qubit QFT
# ==========================================
qft_qc = QuantumCircuit(3)
# Append the primitive gate and immediately decompose it so the UI can draw the universal gates
qft_qc.append(QFTGate(3), [0, 1, 2])
qft_qc = qft_qc.decompose()

db.add_algorithm("Quantum Fourier Transform (3Q)", qft_qc)

# ==========================================
# Add a Challenge: Grover's Search (Target state |101>)
# ==========================================
grover_qc = QuantumCircuit(3)
# Step 1: Initialization to equal superposition
grover_qc.h([0, 1, 2])

# Step 2: Define the Oracle for |101> (State 5)
oracle = QuantumCircuit(3)
oracle.cz(0, 2) # A simple oracle for 101 phase flip

# Step 3: Apply Grover Operator using the updated function
grover_op = grover_operator(oracle)
grover_qc = grover_qc.compose(grover_op)

db.add_challenge_from_circuit("Search Challenge: Amplify |101⟩", grover_qc)
```
