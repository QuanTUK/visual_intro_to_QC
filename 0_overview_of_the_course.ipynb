# A Visual Introduction to Quantum Communication and Computation


Quantum technologies are at the forefront of technological interest, be it for economic purposes or fighting climate change by more efficient simulations and computational models, or advancing science or military endeavours by more accurate sensing techniques or secure communication. They are made up of three pillars: quantum sensing, quantum computation & simulation, and quantum communication. The qubit is the fundamental unit of quantum information and sees applications in all of these pillars. Unlike the classical bit, which is deterministically constrained to a binary state of 0 or 1, the qubit exists as a two-level quantum mechanical system. It is defined not merely by its basis states, but by its ability to exist in a linear coherent superposition of those states. This property, combined with the capacity for entanglement between multiple qubits, marks a departure from classical probability theory and forms the basis of the quantum advantage.

While qubits can be physically realized through diverse substrates—ranging from superconducting circuits and trapped ions to the polarization states of single photons—this course abstracts away the hardware implementation. We treat the qubit as a mathematical information carrier to focus on the logic of quantum protocols, while using visualization rigorously to gain an intuition for the underlying principles.

Standard introductions to Quantum Information Theory often rely exclusively on algebraic formalism. While the mathematics of complex vector spaces is indispensable, it can obscure physical intuition. This course adopts a distinct methodology: using visualization to bridge the gap between abstract linear algebra and physical sense-making. These visualizations are not merely illustrative metaphors; they are exact representations of the quantum state space (such as the Bloch sphere) and circuit logic. They are designed to scale, allowing us to analyze not just single-qubit coherence, but the non-local correlations found in multi-qubit systems.

This series acts as a comprehensive entry point for researchers and students. The content is structured to build complexity layer by layer:

1. Single-Qubit Foundations: We begin by defining the qubit and the principle of superposition. You will learn to visualize quantum states and manipulate them using single qubit operations, establishing the visual intuition required for later protocols.
2. Quantum Security: We apply these foundational principles to secure quantum communication. Specifically, we examine the BB84 Protocol, demonstrating how the fundamental uncertainty principles of quantum mechanics allow for the detection of eavesdroppers in key distribution.
3. Entanglement and non-locality: We then expand the system to two qubits to introduce entanglement, a correlation that no classical theory allows. This section explores the Bell experiment, proving that local hidden variable theories cannot explain quantum mechanics—a result that underpins the E91 Protocol for entanglement-based, secure key distribution.
4. Information Protocols: Finally, we utilize entanglement as a resource. You will derive and visualize protocols that have no classical equivalent, such as superdense coding (transmitting two classical bits via one qubit) and quantum teleportation (transferring quantum states across distances).

The course concludes by briefly looking forward, discussing how principles used in these communication protocols serve as the primitives for universal quantum computation. In order to understand the content, it is helpful that you have a grasp of foundational geometry (sine and cosine), complex numbers (Absolute values and Euler formula), and elementary linear algebra (mostly vector and matrix multiplication).
